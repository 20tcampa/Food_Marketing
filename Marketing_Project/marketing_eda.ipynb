```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('cleaned_df.csv')

df.drop('Unnamed: 0', axis=1, inplace=True)
pd.set_option('display.max_columns', None)
```

```python
df.head()
```

```python
df.describe()
```

__Product Information__

```python
fig, axs = plt.subplots(1,2, figsize=(14,5), layout='tight')

products = df[['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']]
channels = df[['NumDealsPurchases', 'NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases', 'NumWebVisitsMonth']]

sns.barplot(products.sum(), orient='y', ax=axs[0])
axs[0].grid(axis='x')
axs[0].set_title('Quantities Sold')
sns.barplot(channels.sum(), orient='y', ax=axs[1])
axs[1].grid(axis='x')
axs[1].set_title('Frequency of Channels')

plt.show()
```

Things to point out about Product Information:

- Wine and Meat Products appear to be the thing people buy the most.
- In-Store purchases seem to be the most popular option for consumers. 
- There are a higher number of website visits than there are website purchases.

__Demographic Information__

```python
fig, axs = plt.subplots(1,3, figsize=(15,5), layout='tight')

sns.histplot(df['Age'], ax=axs[0]).set_xlabel('')
axs[0].set_title('Distribution of Age')
axs[0].grid(axis='y')
sns.histplot(df['Days_As_Customer'], ax=axs[1]).set_xlabel('')
axs[1].set_title('Distribution of Number of Days as Customer')
axs[1].grid(axis='y')
sns.histplot(df['Recency'], ax=axs[2]).set_xlabel('')
axs[2].set_title('Distribution of Days since last purchase')
axs[2].grid(axis='y')

plt.show()
```

```python
fig, axs = plt.subplots(1,2, figsize=(14,5), layout='tight', sharey=True)

sns.barplot(df, x='Marital_Status', y='Income', hue='Marital_Status', ax=axs[0]).set_xlabel('')
axs[0].set_title('Average Income by Marital Status')
axs[0].grid(axis='y')
sns.barplot(df, x='Education', y='Income', hue='Education', ax=axs[1]).set_xlabel('')
axs[1].set_title('Average Income by Education')
axs[1].grid(axis='y')

plt.show()
```

```python
fig, axs = plt.subplots(1,2, figsize=(15,5), layout='tight', sharey=True)

df.groupby(by='Marital_Status')[['MntWines', 'MntFruits',
       'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts',
       'MntGoldProds', 'Total_Purchases']].mean().plot(kind='bar', ax=axs[0], xlabel='', ylabel='Amount Purchased')
axs[0].legend(bbox_to_anchor=(1.0, 1.0))
axs[0].grid(axis='y')
axs[0].set_title('Products Purchases by Marital Status')

df.groupby(by='Education')[['MntWines', 'MntFruits',
       'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts',
       'MntGoldProds', 'Total_Purchases']].mean().plot(kind='bar', ax=axs[1], xlabel='', ylabel='Amount Purchased')
axs[1].legend(bbox_to_anchor=(1.0, 1.0))
axs[1].grid(axis='y')
axs[1].set_title('Products Purchases by Education')

plt.show()
```

```python
fig, axs = plt.subplots(1,2, figsize=(15,5), layout='tight', sharey=True)

df.groupby('Marital_Status')[['Kidhome','Teenhome']].sum().plot(kind='bar', ax=axs[0], xlabel='', ylabel='Count of Dependets')
axs[0].set_title('Kids/Teens by Marital Status')
axs[0].grid(axis='y')
df.groupby('Education')[['Kidhome','Teenhome']].sum().plot(kind='bar', ax=axs[1], xlabel='', ylabel='Count of Dependets')
axs[1].set_title('Kids/Teens by Education')
axs[1].grid(axis='y')

plt.show()
```

Things to point out from Demographic Information:
- Average Income by Marital Status seems to be roughly the same with Widows having slightly higher income on average.
- Average Income by Education shows that those with a Basic degree have significantly lower income on average.
- Number of products purchased by Marital Status shows that Widows buy significantly more products, particularily more wine. 
- The proportions of product purchased is pretty much the same across Marital Status.
- Number of products purchased by Education shows the same proportions as Marital Status. 
- Those with a Basic Education buy significantly less than others. 
- Those with 2n Cycle buy less products compared to the other Higher-Ed customers.
- Those with an Undergraduate degree have significantly more dependents.
- Widows and those with a Basic education have significantly less dependents.

__Campaign Information__

```python
cmp_cols = ['AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5', 'Response', 'AcceptedAnyCmp', 'Total_AcceptedCmp']

fig, axs = plt.subplots(2,4, figsize=(15,8), layout='tight', sharey=True)

for ax, title in zip(axs.flat, cmp_cols):
    sns.countplot(df, x=title, hue=title, ax=ax, stat='percent').set(xlabel=None)
    ax.set_title(title)
    ax.grid(axis='y')
```

```python
cmp = ['AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5', 'Response', 'AcceptedAnyCmp']

for num, i in enumerate(cmp):
    percent = ((df[i] == 1).sum() / len(df)) * 100
    percent = round(percent, 2)
    if i == 'AcceptedAnyCmp':
        print(f'The success rate of all campaigns is {percent}%')
    else:
        print(f'The success rate of Campaign {num + 1} is {percent}%')
```

```python
accepted_df = df.loc[df['AcceptedAnyCmp'] == 1]
not_accepted_df = df.loc[df['AcceptedAnyCmp'] == 0]
```

Going to see Product/Channel Demographics change according to whether or not a campaign was successful.

```python
products = ['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']
channels = ['NumDealsPurchases', 'NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases', 'NumWebVisitsMonth']

def percent_change(before_df, after_df, columns):
    for i in columns:
        p_change = ((after_df[i].mean() - before_df[i].mean()) / after_df[i].mean()) * 100
        p_change = round(p_change, 2)
        print(f'{i} changed by {p_change}%')
```

```python
fig, axs = plt.subplots(1,2, figsize=(15,5), layout='tight', sharey=True)

sns.barplot(not_accepted_df[channels].mean(), orient='x', ax=axs[0])
axs[0].tick_params(axis='x', rotation=20)
axs[0].grid(axis='y')
axs[0].set_title('Channel Activity for Unsuccessful Campaign')

sns.barplot(accepted_df[channels].mean(), orient='x', ax=axs[1])
axs[1].tick_params(axis='x', rotation=20)
axs[1].grid(axis='y')
axs[1].set_title('Channel Activity for Successful Campaign')

plt.show()
```

```python
percent_change(not_accepted_df, accepted_df, channels)
```

```python
fig, axs = plt.subplots(1,2, figsize=(15,5), layout='tight', sharey=True)

sns.barplot(not_accepted_df[products].mean(), orient='x', ax=axs[0])
axs[0].tick_params(axis='x', rotation=20)
axs[0].grid(axis='y')
axs[0].set_title('Product Activity for Unsuccessful Campaign')

sns.barplot(accepted_df[products].mean(), orient='x', ax=axs[1])
axs[1].tick_params(axis='x', rotation=20)
axs[1].grid(axis='y')
axs[1].set_title('Product Activity for Successful Campaign')

plt.show()
```

```python
percent_change(not_accepted_df, accepted_df, products)
```

Things to note about Campaign Information:

- Regarding campaign results:
    - All campaigns together had a total success rate of 27.56%.
    - The most successful campaign was Campaign 6 with a 15.19% success rate.
    - The least successful campaign was Campaign 2 with a 1.27% success rate.
- Regarding the effects of a successful campaign:
    - Successful campaigns showed to increase in amount of products sold for all products by 37% - 59%. Highest % change was with Wine (59.57%) and Meat (53.96%).
    - Successful campaigns showed to increase purchases from Store, Catalog, and Website by 16.97%, 48.59%, and 26.87% respectively.
- Customer demographics of successful campaigns:
    - Successful campaigns worked more frequently on customers that:
        - Have been customers for a longer period of time.
        - Purchased products more recently.
        - Are single.
        - Have a PhD.
        - Have less Kids/Teens in their home.
